```python
import matplotlib.pyplot as pt
x=[1,2,3,4,5]
y=[7,5,6,3,8]
z=[1,8,4,7,5]
pt.plot(x,y,color="red",label="March")
pt.plot(x,z,label="April")
pt.legend()
pt.show()
```

```python
import numpy as np
import matplotlib.pyplot as pt
x=np.arange(0,360)
y=np.sin(x*np.pi/180)
pt.xlim(0,360)
pt.ylim(-1.2,1.2)
pt.title("Sin Wave")
pt.xlabel("Degree")
pt.ylabel("Value")
pt.plot(x,y)
pt.show()
```
